## Part 3: Manipulating the Data
    
In Part 3, we are going to change columns of the data to more readable ones, switch columns to more understandable ones, set name as index for rows, change the tokens for companies and type to the real names of companies and types and remove the null data.

```python
Master_DataFrame["COMPANY"]=Master_DataFrame["COMPANY"].replace(["A","G","K","N","P","Q","R"],["American Home Food Products","General Mills","Kelloggs","Nabisco","Post","Quaker Oats","Ralston Purina"])
Master_DataFrame["TYPE"]=Master_DataFrame["TYPE"].replace(["C","H"],["Cold","Hot"])
```

* Display a bar chart showing the amount of protein per serving for all cereals, sorted from most protein to least protein.

```python
import matplotlib.pyplot as plot
# create a figure and axis
figure, axis = plot.subplots(figsize=(16,8))

# sort the data from most protein cereals
Master_DataFrame_Sorted=Master_DataFrame.sort_values("PROTEIN",ascending=False)
Master_DataFrame_Sorted=Master_DataFrame_Sorted.reset_index()
Master_DataFrame_Sorted.head(10)
# get the x and y data

# y axis data

name=Master_DataFrame_Sorted["NAME"]

# x axis data
protein = Master_DataFrame_Sorted["PROTEIN"]

# create a horizontal bar chart since the vertical one cannot display the names well
axis.bar(name,protein)

# set the title and labels
axis.set_title ("Amount of Protein Per Serving For Cereals")
plot.xticks(rotation=-90)
#axis.set_xlabel ("Amount of Protein")
axis.set_ylabel ("Name")
```
